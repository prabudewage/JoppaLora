```python
import pygrib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
# Grib (.grib) format data of weather saved in local machine.
path = "/Users/prabu/Desktop/Joppa/Raw_data/Weather/2022/1/1/00"

grbs = pygrib.open(path)
```

```python
# Reading latitudes and longitudes from index-1 (temperature data)
coord = grbs.select()[1]

# Reading U and V components of wind
u_wind = grbs.select()[2].values
v_wind = grbs.select()[3].values
```

```python
# Assigning latitudes and longitudes of wind data into two variables
lats, lons = coord.latlons()
```

```python
# The location of LoRa-53
lon_50 = -96.7578
lat_50 = 32.9924

lon_53 = -96.7577
lat_53 = 32.9923
```

```python
plt.figure(figsize=(15,15))
ax = plt.subplot(projection=ccrs.PlateCarree())

# Plot wind data
plt.quiver(lons, lats, u_wind, v_wind, pivot='middle')

# Plot Sensor locations
lora50 = ax.scatter(lon_50,lat_50, color = 'blue', s=400, alpha=0.5)
lora53 = ax.scatter(lon_53,lat_53, color = 'red', s=400, alpha=0.5)

ax.legend([lora50, lora53], ['LoRa-50 South', 'LoRa-53 North'])
#c = ax.scatter(lo,la, color = col, s=400, alpha=0.5)


# Saving figure in local machine
#plt.savefig("/Users/prabu/Desktop/wind.png", dpi=300)
```
