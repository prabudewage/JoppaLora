```python
import pygrib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
date_time = "2022/4/16/23"
```

```python
# Grib (.grib) format data of weather saved in local machine.
path = "/Users/prabu/Desktop/Joppa/Raw_data/Weather_3/" + date_time
grbs = pygrib.open(path)
```

```python
# Reading latitudes and longitudes from index-1 (temperature data)
coord = grbs.select()[1]

# Reading U and V components of wind
u_wind = grbs.select()[2].values
v_wind = grbs.select()[3].values
```

```python
# Assigning latitudes and longitudes of wind data into two variables
lats, lons = coord.latlons()
```

```python
# EDIT THIS PART - LORA 53 AND 54

# The location of LoRa-53
lon_53 = -96.7504
lat_53 = 32.7160

lon_54 = -96.7472
lat_54 = 32.7104

lon_Central = -96.7479
lat_Central = 32.7153
```

```python
plt.figure(figsize=(15,15))
ax = plt.subplot(projection=ccrs.PlateCarree())

# Plot wind data
plt.quiver(lons, lats, u_wind, v_wind, pivot='middle', width=0.002)

# Plot Sensor locations
lora53 = ax.scatter(lon_53, lat_53, color = 'deepskyblue', s=400, alpha=0.8)
lora54 = ax.scatter(lon_54, lat_54, color = 'dodgerblue', s=400, alpha=0.8)
centr = ax.scatter(lon_Central, lat_Central, color = 'lime', s=400, alpha=0.8)

ax.legend([lora53, lora54, centr], ['LoRa-53 North', 'LoRa-54 South', 'Central-Node'])
plt.title("Wind Data on " + date_time)
#c = ax.scatter(lo,la, color = col, s=400, alpha=0.5)

date_time = date_time.replace("/", "-")
#tes = "/Users/prabu/Desktop/Joppa/wind_plot/wind_on_" + date_time + ".png"
# Saving figure in local machine
plt.savefig("/Users/prabu/Desktop/Joppa/wind_plot/wind_on_" + date_time + ".png", dpi=300)
```
