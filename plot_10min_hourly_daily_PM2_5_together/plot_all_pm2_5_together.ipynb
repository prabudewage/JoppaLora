# Plot PM2.5 data from all three sensors together for available time period

```julia
using Plots
using CSV
using DataFrames
using Dates
using Statistics
using StatsPlots
```

### Plot 10min averaged PM2.5 data

```julia
# Open local time 10min averaged dataFrames
df53 = CSV.read("/Users/prabu/Desktop/Joppa/local_10min_avg_df/df53_local_10avg_w_dayName.csv", DataFrame)
df54 = CSV.read("/Users/prabu/Desktop/Joppa/local_10min_avg_df/df54_local_10avg_w_dayName.csv", DataFrame);
df_central = CSV.read("/Users/prabu/Desktop/Joppa/local_10min_avg_df/df_central_local_10avg_w_dayName.csv", 
    DataFrame);
```

```julia
# Plot 10min averaged PM2.5 data
p = plot(df53.dateTime, df53.IPS7100_pm2_5, ylim=(0, 200), label = "LoRa 53 (North)", dpi=300)
plot!(df54.dateTime, df54.IPS7100_pm2_5, ylim=(0, 200), label = "Node 54 (South)", dpi=300)
plot!(df_central.dateTime, df_central.IPS7100_pm2_5, ylim=(0, 200), label = "Central Node", dpi=300)
title!("10 Minutes Average PM2.5")
xlabel!("Local Time")
ylabel!("PM2.5 (ug/m3)")
display(p)

savefig("/Users/prabu/Desktop/Joppa/outPut/local_10min_avg_pm2_5_all.png")
```

### Plot hourly averaged PM2.5 data

### Plot daily averaged PM2.5 data

```julia
# Open local time daily averaged dataFrames
df53_d = CSV.read("/Users/prabu/Desktop/Joppa/local_daily_avg_df/df53_local_daily_avg_w_dayName.csv", DataFrame)
df54_d = CSV.read("/Users/prabu/Desktop/Joppa/local_daily_avg_df/df54_local_daily_avg_w_dayName.csv", DataFrame);
df_central_d = CSV.read("/Users/prabu/Desktop/Joppa/local_daily_avg_df/df_central_local_daily_avg_w_dayName.csv", 
    DataFrame);
```

```julia
# Plot daily averaged PM2.5 data
p_d = plot(df53_d.dateTime, df53_d.IPS7100_pm2_5_mean, ylim=(0, 100), label = "LoRa 53 (North)", dpi=300)
plot!(df54_d.dateTime, df54_d.IPS7100_pm2_5_mean, ylim=(0, 100), label = "Node 54 (South)", dpi=300)
plot!(df_central_d.dateTime, df_central_d.IPS7100_pm2_5_mean, ylim=(0, 100), label = "Central Node", dpi=300)
title!("Daily Average PM2.5")
xlabel!("Local Time")
ylabel!("PM2.5 (ug/m3)")
display(p_d)

savefig("/Users/prabu/Desktop/Joppa/outPut/local_daily_avg_pm2_5_all.png")
```

```julia
# adjuesting x axis tick
t1 = df53_d.dateTime[1]                              # Starting time
t2 = df53_d.dateTime[length(df53_d.dateTime)]        # Ending time
tick = t1:Day(2):t2                                  # Time array with 5 hour gap
```

```julia
# Plot daily averaged PM2.5 data
p_d = plot(df53_d.dateTime, df53_d.IPS7100_pm2_5_mean, xticks = tick, xrotation = 90, ylim=(0, 100), 
    label = "LoRa 53 (North)", dpi=300)
plot!(df54_d.dateTime, df54_d.IPS7100_pm2_5_mean, ylim=(0, 100), label = "Node 54 (South)", dpi=300)
plot!(df_central_d.dateTime, df_central_d.IPS7100_pm2_5_mean, ylim=(0, 100), label = "Central Node", dpi=300)
title!("Daily Average PM2.5")
xlabel!("Local Time")
ylabel!("PM2.5 (ug/m3)")
display(p_d)

savefig("/Users/prabu/Desktop/Joppa/outPut/local_daily_avg_pm2_5_all.png")
```
