# Scatter Plot of PM2.5 vs Week Number with align dayNames. 

```julia
using Plots
using CSV
using DataFrames
using Dates
using Statistics
using StatsPlots
```

```julia
# Open local time daily averaged dataFrames
df53_d = CSV.read("/Users/prabu/Desktop/Joppa/local_daily_avg_df/df53_local_daily_avg_w_dayName.csv", DataFrame)
df54_d = CSV.read("/Users/prabu/Desktop/Joppa/local_daily_avg_df/df54_local_daily_avg_w_dayName.csv", DataFrame);
df_central_d = CSV.read("/Users/prabu/Desktop/Joppa/local_daily_avg_df/df_central_local_daily_avg_w_dayName.csv", 
    DataFrame);
```

### Group dataFrame by day name

```julia
weekdays_group_53 = groupby(df53_d, :day_name)
weekdays_group_54 = groupby(df54_d, :day_name) 
weekdays_group_central = groupby(df_central_d, :day_name);
```

### Covert each group into dataFrame 

```julia
# LoRa 53
satL53 = DataFrame(weekdays_group_53[1])
sunL53 = DataFrame(weekdays_group_53[2])
monL53 = DataFrame(weekdays_group_53[3])
tueL53 = DataFrame(weekdays_group_53[4])
wedL53 = DataFrame(weekdays_group_53[5])
thuL53 = DataFrame(weekdays_group_53[6])
friL53 = DataFrame(weekdays_group_53[7])

# LoRa 54
satL54 = DataFrame(weekdays_group_54[1])
sunL54 = DataFrame(weekdays_group_54[2])
monL54 = DataFrame(weekdays_group_54[3])
tueL54 = DataFrame(weekdays_group_54[4])
wedL54 = DataFrame(weekdays_group_54[5])
thuL54 = DataFrame(weekdays_group_54[6])
friL54 = DataFrame(weekdays_group_54[7])

# Central Node
satCent = DataFrame(weekdays_group_central[1])
sunCent = DataFrame(weekdays_group_central[2])
monCent = DataFrame(weekdays_group_central[3])
tueCent = DataFrame(weekdays_group_central[4])
wedCent = DataFrame(weekdays_group_central[5])
thuCent = DataFrame(weekdays_group_central[6])
friCent = DataFrame(weekdays_group_central[7]);
```

### Align dates of same week

```julia
# All 7 days of a week should be align
# Now first day of the combine dataFrame is Saturday
# So, reduce 1 day from Sunday and 2 days from Monday,.. etc to align with Saturday

# LoRa53
sunL53.dateTime = sunL53.dateTime - Dates.Day(1) 
monL53.dateTime = monL53.dateTime - Dates.Day(2)
tueL53.dateTime = tueL53.dateTime - Dates.Day(3)
wedL53.dateTime = wedL53.dateTime - Dates.Day(4)
thuL53.dateTime = thuL53.dateTime - Dates.Day(5)
friL53.dateTime = friL53.dateTime - Dates.Day(6);

# LoRa54
sunL54.dateTime = sunL54.dateTime - Dates.Day(1) 
monL54.dateTime = monL54.dateTime - Dates.Day(2)
tueL54.dateTime = tueL54.dateTime - Dates.Day(3)
wedL54.dateTime = wedL54.dateTime - Dates.Day(4)
thuL54.dateTime = thuL54.dateTime - Dates.Day(5)
friL54.dateTime = friL54.dateTime - Dates.Day(6);

# Central Node
sunCent.dateTime = sunCent.dateTime - Dates.Day(1) 
monCent.dateTime = monCent.dateTime - Dates.Day(2)
tueCent.dateTime = tueCent.dateTime - Dates.Day(3)
wedCent.dateTime = wedCent.dateTime - Dates.Day(4)
thuCent.dateTime = thuCent.dateTime - Dates.Day(5)
friCent.dateTime = friCent.dateTime - Dates.Day(6);
```

### Scatter plot of PM2.5 with week number

```julia
# LoRa53
L53 = plot(satL53.dateTime, satL53.IPS7100_pm2_5_mean, seriestype = :scatter, ylim=(0, 75), 
    label = "Saturday", dpi=300)
plot!(sunL53.dateTime, sunL53.IPS7100_pm2_5_mean, seriestype = :scatter, label = "Sunday", dpi=300)
plot!(monL53.dateTime, monL53.IPS7100_pm2_5_mean, seriestype = :scatter, label = "Monday", dpi=300)
plot!(tueL53.dateTime, tueL53.IPS7100_pm2_5_mean, seriestype = :scatter, label = "Tuesday", dpi=300)
plot!(wedL53.dateTime, wedL53.IPS7100_pm2_5_mean, seriestype = :scatter, label = "Wednesday", dpi=300)
plot!(thuL53.dateTime, thuL53.IPS7100_pm2_5_mean, seriestype = :scatter, label = "Thursday", dpi=300)
plot!(friL53.dateTime, friL53.IPS7100_pm2_5_mean, seriestype = :scatter, label = "Friday", dpi=300)
title!("LoRa53 - PM2.5 vs Week Number")
xlabel!("Local Date")
ylabel!("PM2.5 (ug/m3)")
display(L53)
 
savefig("/Users/prabu/Desktop/Joppa/outPut/LoRa53_daily_avg_PM2.5_with_align_dayname.png")
```

# Plot PM2.5 vs DayName bar chart.

### Use the original dataFrame > Group by dayName > Average each group > Combine to a new dataFrame

```julia
# LoRa53
L53_sub = select(df53_d, [:day_name, :IPS7100_pm2_5_mean]) 
weekdays_group_L53 = groupby(L53_sub, :day_name)
com_L53 = combine(weekdays_group_L53, valuecols(weekdays_group_L53) .=> mean)

# LoRa53
L54_sub = select(df54_d, [:day_name, :IPS7100_pm2_5_mean]) 
weekdays_group_L54 = groupby(L54_sub, :day_name)
com_L54 = combine(weekdays_group_L54, valuecols(weekdays_group_L54) .=> mean)

# Central Node
Cent_sub = select(df_central_d, [:day_name, :IPS7100_pm2_5_mean]) 
weekdays_group_Cent = groupby(Cent_sub, :day_name)
com_Cent = combine(weekdays_group_Cent, valuecols(weekdays_group_Cent) .=> mean);
```

### Plot bar charts

```julia
# LoRa53
p_53 = bar(com_L53.day_name, com_L53.IPS7100_pm2_5_mean_mean, ylim=(0, 20), labels=false)
title!("LoRa53 - PM2.5 with Day Names")
xlabel!("Day Name")
ylabel!("PM2.5 (ug/m3)")
```

```julia
# LoRa54
p_54 = bar(com_L54.day_name, com_L54.IPS7100_pm2_5_mean_mean, ylim=(0, 25), labels=false)
title!("LoRa54 - PM2.5 with Day Names")
xlabel!("Day Name")
ylabel!("PM2.5 (ug/m3)")
```

```julia
# Central Node
p_Cent = bar(com_Cent.day_name, com_Cent.IPS7100_pm2_5_mean_mean, ylim=(0, 15), labels=false)
title!("Central Node - PM2.5 with Day Names")
xlabel!("Day Name")
ylabel!("PM2.5 (ug/m3)")
display(p_Cent)
 
savefig("/Users/prabu/Desktop/Joppa/outPut/CentralN_daily_avg_PM2.5_with_dayname_bar.png")
```

### Adding WHO PM2.5 guideline values

#### According to WHO (https://www.who.int/news-room/fact-sheets/detail/ambient-(outdoor)-air-quality-and-health) PM2.5 concenration should be less than 5 μg/m3 annual mean and 15 μg/m3 24-hour mean

#### Make a new column (above) with exceed PM2.5 value than 15.
#### Make 0 all negatives in that above column.
#### Make a new column (below) with PM2.5 less than 15.

```julia
# LoRa53
com_L53[!,:above] = com_L53.IPS7100_pm2_5_mean_mean .- 15           
com_L53."above" = ifelse.(com_L53."above" .< 0, 0, com_L53."above")  
com_L53[!,:below] = com_L53.IPS7100_pm2_5_mean_mean .- com_L53.above 

# LoRa54
com_L54[!,:above] = com_L54.IPS7100_pm2_5_mean_mean .- 15           
com_L54."above" = ifelse.(com_L54."above" .< 0, 0, com_L54."above")  
com_L54[!,:below] = com_L54.IPS7100_pm2_5_mean_mean .- com_L54.above;
```

#### Plot the bar chart

```julia
# LoRa53
bar53 = groupedbar(hcat(com_L53.above, com_L53.below), bar_position=:stack, color=["tomato" "deepskyblue"], 
    labels=false, dpi=300)
title!("LoRa53 - PM2.5 with WHO Guideline")
xlabel!("Day Name")
ylabel!("PM2.5 (ug/m3)")
display(bar53)
 
savefig("/Users/prabu/Desktop/Joppa/outPut/L53_daily_avg_PM2.5_with_dayname_bar.png")
```

```julia
# LoRa54
bar54 = groupedbar(hcat(com_L54.above, com_L54.below), bar_position=:stack, color=["tomato" "deepskyblue"], 
    labels=false, dpi=300)
title!("LoRa54 - PM2.5 with WHO Guideline")
xlabel!("Day Name")
ylabel!("PM2.5 (ug/m3)")
display(bar54)
 
savefig("/Users/prabu/Desktop/Joppa/outPut/L54_daily_avg_PM2.5_with_dayname_bar.png")
```
